3) (a) Using Dataset-2


```python
# uploading the training data(dataset-2)
uploaded_train = files.upload()
```

```python
# uploading the testing data(dataset-2)
uploaded_test = files.upload()
```

```python
# reading train_data from training data set(dataset-2)

df1 = pd.read_csv(io.BytesIO(uploaded_train['train_41_ds_2.csv']))

train_data = df1.to_numpy()
train_data = train_data.transpose()
print(train_data)
```

```python
# reading test_data from testing data set(dataset-2)

df2 = pd.read_csv(io.BytesIO(uploaded_test['test_41_ds_2.csv']))

test_data = df2.to_numpy()
test_data = test_data.transpose()
print(test_data)
```

```python
# getting the values of each coordinate of training and testing data in different variables

X_train = np.array(train_data[0])
Y_train = np.array(train_data[1])
label_train = train_data[2]

X_test = np.array(test_data[0])
Y_test = np.array(test_data[1])
label_test = test_data[2]


# finding the data points(training and testing) for each class and plotting
indices_train_0 = [i for i in range(len(label_train)) if label_train[i]==0]
indices_train_1 = [i for i in range(len(label_train)) if label_train[i]==1]
indices_train_2 = [i for i in range(len(label_train)) if label_train[i]==2]

indices_test_0 = [i for i in range(len(label_test)) if label_test[i]==0]
indices_test_1 = [i for i in range(len(label_test)) if label_test[i]==1]
indices_test_2 = [i for i in range(len(label_test)) if label_test[i]==2]

X_train_0 = X_train[indices_train_0]
Y_train_0 = Y_train[indices_train_0]

X_train_1 = X_train[indices_train_1]
Y_train_1 = Y_train[indices_train_1]

X_train_2 = X_train[indices_train_2]
Y_train_2 = Y_train[indices_train_2]


X_test_0 = X_test[indices_test_0]
Y_test_0 = Y_test[indices_test_0]

X_test_1 = X_test[indices_test_1]
Y_test_1 = Y_test[indices_test_1]

X_test_2 = X_test[indices_test_2]
Y_test_2 = Y_test[indices_test_2]


# plotting
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

ax1.scatter(X_train_0, Y_train_0,label = "0")
ax1.scatter(X_train_1, Y_train_1,label = "1")
ax1.scatter(X_train_2, Y_train_2,label = "2")
ax1.set_title('Training data')
ax2.scatter(X_test_0, Y_test_0,label = "0")
ax2.scatter(X_test_1, Y_test_1,label = "1")
ax2.scatter(X_test_2, Y_test_2,label = "2")
ax2.set_title('Testing data')

fig.suptitle('Scatter Plots')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax1.legend()
ax2.legend()

plt.show()
```

Seperating the training set of dataset-2 into training and validation set.

```python
# seperating the training and validation data

validation_data_ds_2 = train_data[:,math.ceil(train_data.shape[1]*0.8):]
train_data_1_ds_2 = train_data[:,:math.ceil(train_data.shape[1]*0.8)]

X_validation_ds_2 = np.array(validation_data_ds_2[0])
Y_validation_ds_2 = np.array(validation_data_ds_2[1])
label_validation_ds_2 = validation_data_ds_2[2]

X_train = np.array(train_data_1_ds_2[0])
Y_train = np.array(train_data_1_ds_2[1])
label_train = train_data_1_ds_2[2]
```

```python
# defining the KNN classifier class and also defining it's methods

class KNN_classifier:

  def __init__(self, k=3):
        self.k = k

  def fit(self, X, y, label):
        self.X_train = X
        self.Y_train = y
        self.label_train = label

  def predict(self, X_test,Y_test):

        pred_label = []

        for i in range(len(X_test)):

          n = len(self.X_train)
          
          X_test_l = np.ones(n)*X_test[i]
          Y_test_l = np.ones(n)*Y_test[i]

          matrix = np.zeros((n,2))

          distances = np.sqrt((X_test_l-self.X_train)**2 + (Y_test_l - self.Y_train)**2)

          matrix[:,0] = distances
          matrix[:,1] = self.label_train

          indices = matrix[:, 0].argsort()
          sorted_matrix = matrix[indices]

          knn_y = sorted_matrix[:self.k,1]

          max_count = 0
          max_count_element = -100
          for i in range(len(knn_y)):
            count = 0
            for j in range(len(knn_y)):
              if knn_y[i] == knn_y[j]:
                count += 1
            
            if count > max_count :
              max_count = count
              max_count_element = knn_y[i]
          
          pred_label+=[max_count_element]
      
        return pred_label



```

3(a)(i) Finding the optimal value of k using the validation set.

```python
accuracy_list=[]
k_list=[]
for k in range(1,21):

  knn = KNN_classifier(k)

  knn.fit(X_train, Y_train,label_train)

  pred_label_validation = knn.predict(X_validation_ds_2,Y_validation_ds_2)     # predicting values for the validation set

  error=pred_label_validation - label_validation_ds_2
  accuracy=np.count_nonzero(error==0)/len(error)
  k_list.append(k)
  accuracy_list.append(accuracy)

best_k =k_list[np.argmax(np.array(accuracy_list))]

# plotting accuracy vs k for the validation data
fig, ax = plt.subplots()

plt.plot(k_list,accuracy_list)
plt.xticks(k_list)

fig.suptitle('accuracy vs k')
ax.set_xlabel('k')
ax.set_ylabel('accuracy')

plt.show()
```

From the above accuracy vs k graph we see that from k = 1 to k = 20, all give an accuracy of 1 for the test data. So we will use k = 1 as our best model.

3) a) ii) Finding the accuracy on train and test data for the best model(k=1).

```python
knn = KNN_classifier(best_k)

knn.fit(X_train, Y_train,label_train)


# accuracy on train data
pred_label_train = knn.predict(X_train,Y_train)

error_train=pred_label_train - label_train
accuracy_train=np.count_nonzero(error_train==0)/len(error_train)

# accuracy on test data
pred_label_test = knn.predict(X_test,Y_test)

error_test=pred_label_test - label_test
accuracy_test=np.count_nonzero(error_test==0)/len(error_test)


print("accuracy of best model on train data of dataset 2 = ",accuracy_train)
print("accuracy of best model on test data of dataset 2 = ",accuracy_test)
```

3(a)(iii) The test data along with the classification boundary for the best model(k=1) is being plotted.

```python
k = best_k

knn = KNN_classifier(k)

knn.fit(X_train, Y_train,label_train)

# plotting the classification boundary

# generating all points in the given range
class_boundary_X = []
class_boundary_Y = []

for x in np.arange(-30,12,0.6):
  for y in np.arange(-30,40,1):
    class_boundary_X += [x]
    class_boundary_Y += [y]

class_boundary_Y = np.array(class_boundary_Y)
class_boundary_X = np.array(class_boundary_X[:len(class_boundary_Y)])

# predicting the class of all the generated points
pred_label_test_ds_2 = knn.predict(class_boundary_X,class_boundary_Y)

# finding the data points of the generated points of each class
indices_0 = [i for i in range(len(pred_label_test_ds_2)) if pred_label_test_ds_2[i]==0]
indices_1 = [i for i in range(len(pred_label_test_ds_2)) if pred_label_test_ds_2[i]==1]
indices_2 = [i for i in range(len(pred_label_test_ds_2)) if pred_label_test_ds_2[i]==2]

class_boundary_X_0 = class_boundary_X[indices_0]
class_boundary_Y_0 = class_boundary_Y[indices_0]

class_boundary_X_1 = class_boundary_X[indices_1]
class_boundary_Y_1 = class_boundary_Y[indices_1]

class_boundary_X_2 = class_boundary_X[indices_2]
class_boundary_Y_2 = class_boundary_Y[indices_2]


# plotting the test points and the classification boundary
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))

ax.scatter(class_boundary_X_0, class_boundary_Y_0,label="0")
ax.scatter(class_boundary_X_1,class_boundary_Y_1,label="1")
ax.scatter(class_boundary_X_2, class_boundary_Y_2,label="2")
ax.scatter(X_test_0, Y_test_0, color=(0, 0, 100/255))
ax.scatter(X_test_1,Y_test_1, color=(155/255, 0, 0))
ax.scatter(X_test_2, Y_test_2, color=(0, 100/255, 0))

ax.set_title('Training data')

fig.suptitle('Using the knn classifier on dataset-2 testing data for k = '+str(k))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()

plt.show()
```

Question 3(a) iv) Finding the confusion matrix for knn(with k = 1) for the training and testing data for dataset-2

```python
# Calculate the confusion matrix for training and testing data
# where rows depict the predicted values and the columns depict the actual values

conf_mat_test = np.zeros((3, 3))
for i in range(len(label_test)):
    conf_mat_test[int(pred_label_test[i]), int(label_test[i])] += 1

conf_mat_train = np.zeros((3,3))
for i in range(len(label_train)):
    conf_mat_train[int(pred_label_train[i]), int(label_train[i])] += 1

print("Confusion matrix of test data :\n", conf_mat_test)
print("Confusion matrix of train data :\n", conf_mat_train)
```
